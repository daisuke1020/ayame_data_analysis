【問題6】図の作成

ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

```python
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。
df_en = np.array([150,150,150,150])
plt.title("iris")
label = ["sepal_length","sepal_width","petal_length","petal_width"]
plt.pie(df_en,labels=label,counterclock=False,startangle=90,autopct="%1.1f%%")  # 「リスト名」のところにリストの変数名を入れる
plt.show()  # com + con + enter
```

特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

```python
sepal_length_list = iris.data[:,0]
sepal_width_list = iris.data[:,1]
petal_length = iris.data[:,2]
petal_width = iris.data[:,3]
iris_list =(sepal_length_list,sepal_width_list,petal_length,petal_width)
```

```python
plt.title("iris")
plt.xticks([1,2,3,4],["sepal_length","sepal_width","petal_length","petal_width"])
plt.ylim([0,8])
plt.xlim([0,5])
plt.grid()
plt.boxplot(iris_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()  # com + con + enter
```

特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化す

```python
plt.title("iris")
plt.xticks([1,2,3,4],["sepal_length","sepal_width","petal_length","petal_width"])
plt.ylim([0,8])
plt.xlim([0,5])
plt.grid()
plt.violinplot(iris_list,showmedians=True)  # 「リスト名」のところにリストの変数名を入れる
plt.show()  # com + con + enter
```

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。

違い
各データの分布がそのまま図に現れるか現れないかの差。

箱ひげ図の利点
単峰性かつ四分位数や外れ値が貴重な場合。

バイオリン図の利点
数値データの分布を可視化、比較したい場合。

【問題7】特徴量間の関係の確認

４つの特徴量から２つ選ぶようなすべての組み合わせ（６つ）の散布図を種類ごとに色分けして作る

```python
#６つの散布図
#sl*sw
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.title("iris")
plt.scatter(sepal_length_list[0:51],sepal_width_list[0:51],label="0")
plt.scatter(sepal_length_list[50:101],sepal_width_list[50:101],label="1")
plt.scatter(sepal_length_list[100:-1],sepal_width_list[100:-1],label="2")
plt.legend()
plt.show()

#sl*pl
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.title("iris")
plt.scatter(sepal_length_list[0:51],petal_length[0:51],label="0")
plt.scatter(sepal_length_list[50:101],petal_length[50:101],label="1")
plt.scatter(sepal_length_list[100:-1],petal_length[100:-1],label="2")
plt.legend()
plt.show()

#sl*pw
plt.xlabel("sepal_length")
plt.ylabel("petal_width")
plt.title("iris")
plt.scatter(sepal_length_list[0:51],petal_width[0:51],label="0")
plt.scatter(sepal_length_list[50:101],petal_width[50:101],label="1")
plt.scatter(sepal_length_list[100:-1],petal_width[100:-1],label="2")
plt.legend()
plt.show()

#sw*pl
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.title("iris")
plt.scatter(sepal_width_list[0:51],petal_length[0:51],label="0")
plt.scatter(sepal_width_list[50:101],petal_length[50:101],label="1")
plt.scatter(sepal_width_list[100:-1],petal_length[100:-1],label="2")
plt.legend()
plt.show()

#sw*pw
plt.xlabel("sepal_width")
plt.ylabel("petal_width")
plt.title("iris")
plt.scatter(sepal_width_list[0:51],petal_width[0:51],label="0")
plt.scatter(sepal_width_list[50:101],petal_width[50:101],label="1")
plt.scatter(sepal_width_list[100:-1],petal_width[100:-1],label="2")
plt.legend()
plt.show()

#pl*pw
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.title("iris")
plt.scatter(petal_length[0:51],petal_width[0:51],label="0")
plt.scatter(petal_length[50:101],petal_width[50:101],label="1")
plt.scatter(petal_length[100:-1],petal_width[100:-1],label="2")
plt.legend()
plt.show()
```
